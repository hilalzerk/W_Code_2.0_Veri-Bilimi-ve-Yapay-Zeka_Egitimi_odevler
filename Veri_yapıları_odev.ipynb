# 3
Kullanıcıdan iki değer girmesini istenir. Girilen değerlerin toplama, çıkarma, çarpma, bölme sonuçlarını yazdırılır.

```python
key = 1 
while key == 1:
    islem = int(input("Yapmak istediğiniz işlemi giriniz: "))

    if islem == 1:
        sayi1 = int(input("İlk sayıyı giriniz: "))
        sayi2 = int(input("İkinci sayıyı giriniz: "))
        sonuc = sayi1 + sayi2
        print("Yaptığınız işlemin sonucu:", sonuc)

    elif islem == 2:
        sayi1 = int(input("İlk sayıyı giriniz: "))
        sayi2 = int(input("İkinci sayıyı giriniz: "))
        sonuc = sayi1 - sayi2
        print("Yaptığınız işlemin sonucu:", sonuc)
               
    elif islem == 3:
        sayi1 = int(input("İlk sayıyı giriniz: "))
        sayi2 = int(input("İkinci sayıyı giriniz: "))
        sonuc = sayi1 * sayi2
        print("Yaptığınız işlemin sonucu:", sonuc)

    elif islem == 4:
        sayi1 = int(input("İlk sayıyı giriniz: "))
        sayi2 = int(input("İkinci sayıyı giriniz: "))
        sonuc = sayi1 / sayi2
        print("Yaptığınız işlemin sonucu:", sonuc)

    elif islem == 0:
        print("Program sonlandırıldı.")
        key = 0
```
